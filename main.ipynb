```python
df = pd.read_csv('bank_enriched_addresses.csv')
```

```python
def ages_into_bins(age):
    if age < 10:
        return 0
    elif age < 20:
        return 1
    elif age < 30:
        return 2
    elif age < 40:
        return 3
    elif age < 50:
        return 4
    elif age < 60:
        return 5
    elif age < 70:
        return 6
    elif age < 80:
        return 7
    elif age < 90:
        return 8
    elif age < 100:
        return 9
    else:
        return 10
```

```python
# remove rows where pdays = -1
df = df[df['pdays'] != -1]

# replace day/month with single date in form dd/MM
# reanmee y column outcome
df = df.rename({'y': 'outcome'}, axis=1)
# cateogrization column of the address
```

```python
# replace day/month with single date in form dd/MM
```

```python
# reanmee y column outcome and date outcome
df = df.rename({'y': 'outcome'}, axis=1)
```

```python
# split name into first name and second name columns, drop name
df['name'].apply(lambda x: len(x.split())).value_counts() # checks that all names have two values
df['first_name'] = df['name'].apply(lambda x: x.split()[0])
df['second_name'] = df['name'].apply(lambda x: x.split()[1])
```

```python
# replace ages with bins
df['age'] = df['age'].apply(lambda x: ages_into_bins(x))
```

```python
# replace yes no with booleans
boolean_columns = ['default', 'housing', 'loan', 'outcome']
for column in boolean_columns:
    # print(df[column].value_counts()) used to ensure that columns were boolean
    df[column] = df[column] == 'yes'
```
